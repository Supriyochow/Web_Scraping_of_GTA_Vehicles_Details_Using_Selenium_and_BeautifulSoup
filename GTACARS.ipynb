```python
import requests
import urllib
import urllib.request
import urllib.parse
from bs4 import BeautifulSoup as bs
import io
from selenium import webdriver
import time
import sys
sys.path.insert(0,'/usr/lib/chromium-browser/chromedriver')
options = webdriver.ChromeOptions()
options.add_argument('--headless')
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')
```

```python
all_urls=[]
link="https://www.gtabase.com/grand-theft-auto-v/vehicles/#sort=attr.ct3.frontend_value&sortdir=desc&page="
for i in range(9):
  link1=link+str(i+1)
  all_urls.append(link1)

print(all_urls)

```

```python
all1_urls=[]
for i in range(len(all_urls)):
  browser=webdriver.Chrome(options=options)
  url=all_urls[i]
  browser.get(url)
  time.sleep(10)
  html=browser.page_source
  data=bs(html,'html.parser')
  read1=data.select('.product-item')  
  for j in range(len(read1)):
    x=read1[j].find_all("a")
    for k in range(len(x)):
      y=x[k].get("href")
      part="https://www.gtabase.com"
      link1=part+y
      all1_urls.append(link1)
      print(part+y)
```

```python
car_dic={}
car_no=0
car_name=""
class_name=""
man_name=""
vec_fea=""
price=""
tspeed=""
seats=""
mass=""
drive=""
gears=""
release=""
speed=""
acc=""
brake=""
handle=""
overall=""
for i in range(len(all1_urls)):
  browser=webdriver.Chrome('chromedriver',options=options)
  url=all1_urls[i]
  browser.get(url)
  #time.sleep(10)
  html=browser.page_source
  data=bs(html,'html.parser')
  read=data.select(".article-content")
  
  #print(read)
  #checking any 404 pages
  #print(len(read))
  check=data.find_all('h2',{'class':'title'})
  if check==[]:
    #collecting data
    for j in range(len(read)):
      x=read[j].find_all('dd',{'class':'field-entry'})
      #print(x)
      
      #print(car_no)
      for k in range(len(x)):
        z=x[k].find('span',{'class':'field-label'})
        #print(z)
        if z is not None:
          type_name=z.get_text()
          #print(type_name)
        else:
          continue  
        
        
        #print(type_name)
        if type_name == "Vehicle Name":
          car_name_check=x[k].find('span',{'class':'field-value'})
          if car_name_check is not None:
            car_name=car_name_check.get_text()

            #print(car_name)
          else:
            car_name="N/A"  

        elif type_name == "Vehicle Class":
          class_name_check=x[k].find('span',{'class':'field-value'})
          if class_name_check is not None:
            class_name=class_name_check.get_text()

            #print(car_name)
          else:
            class_name="N/A"
          #print(class_name) 

        elif type_name == "Manufacturer":
          man_name_check=x[k].find('span',{'class':'field-value'})
          if man_name_check is not None:
            man_name=man_name_check.get_text()

            #print(man_name)
          else:
            man_name="N/A"

        elif type_name == "Vehicle Features":
          vec_fea_check=x[k].find('span',{'class':'field-value'})
          if vec_fea_check is not None:
            vec_fea=vec_fea_check.get_text()

            #print(vec_fea)
          else:
            vec_fea="N/A"

        elif type_name == "Purchase Price":
          price_check=x[k].find('span',{'class':'field-value'})
          if price_check is not None:
            price=price_check.get_text()

            #print(price)
          else:
            price="N/A" 
        elif type_name == "Top Speed (Game Files)":
          tspeed_check=x[k].find('span',{'class':'field-value'})
          if tspeed_check is not None:
            tspeed=tspeed_check.get_text()

            #print(tspeed)
          else:
            tspeed="N/A"

        elif type_name == "Seats":
          seats_check=x[k].find('span',{'class':'field-value'})
          if seats_check is not None:
            seats=seats_check.get_text()

            #print(seats)
          else:
            seats="N/A"

        elif type_name == "Mass / Weight":
          mass_check=x[k].find('span',{'class':'field-value'})
          if mass_check is not None:
            mass=mass_check.get_text()

            #print(mass)
          else:
            mass="N/A"   


        elif type_name == "Drive Train":
          drive_check=x[k].find('span',{'class':'field-value'})
          if drive_check is not None:
            drive=drive_check.get_text()

            #print(drive)
          else:
            drive="N/A"   

        elif type_name == "Gears":
          gears_check=x[k].find('span',{'class':'field-value'})
          if gears_check is not None:
            gears=gears_check.get_text()

            #print(gears)
          else:
            gears="N/A"                                


        elif type_name == "Release Date":
          release_check=x[k].find('span',{'class':'field-value'})
          if release_check is not None:
            release=release_check.get_text()

            #print(release)
          else:
            release="N/A"    
          #print(car_no)

      
      #car_dic['Number']=car_no
      if car_name=="":
        car_name="N/A"
      else:
        pass
      if class_name=="":
        class_name="N/A"
      else:
        pass

      if man_name=="":
        man_name="N/A"
      else:
        pass 

      if price=="":
        price="N/A"
      else:
        pass  
      if tspeed=="":
        tspeed="N/A"
      else:
        pass

      if seats=="":
        seats="N/A"
      else:
        pass  

      if mass=="":
        mass="N/A"
      else:
        pass

      if drive=="":
        drive="N/A"
      else:
        pass  

      if gears=="":
        gears="N/A"
      else:
        pass

      if release=="":
        release="N/A"
      else:
        pass
    for j in range(len(read)):
      x=read[j].find_all('div',{'class':'gta5-stats'})
      for k in range(len(x)):
        y=x[k].find_all('dd',{'class':'field-entry'})
        for l in range(len(y)):
          z=y[l].find('div',{'class':'field-label'})
          if z is not None:
            type_name=z.get_text()
            #print(type_name)
          else:
            continue
          #print(z)
          #print(type_name)
          if type_name == "Speed":
            speed_name_check=x[k].find('div',{'class':'field-value'})
            if speed_name_check is not None:
              speed=speed_name_check.get_text()

              #print(speed)
            else:
              speed="N/A"

          if type_name == "Acceleration":
            acc_name_check=x[k].find_all('div',{'class':'field-value'})
            acc_name1=acc_name_check[1]
            if acc_name_check is not None:
              acc=acc_name1.get_text()

              #print(acc)
            else:
              acc="N/A"     
              #print()   
              
          if type_name == "Braking":
            brake_name_check=x[k].find_all('div',{'class':'field-value'})
            brake_name1=brake_name_check[2]
            if brake_name_check is not None:
              brake=brake_name1.get_text()

              #print(brake)
            else:
              brake="N/A"     
              #print() 

          if type_name == "Handling":
            handle_name_check=x[k].find_all('div',{'class':'field-value'})
            handle_name1=handle_name_check[-1]
            if handle_name_check is not None:
              handle=handle_name1.get_text()

              #print(handle)
            else:
              handle="N/A"     
              #print()

          if type_name == "Speed":
            overall_name_check=x[k].find('span',{'class':'field-value'})
            if overall_name_check is not None:
              overall=overall_name_check.get_text()

              #print(overall)
            else:
              overall="N/A"     
              #print()
    #print(0)          
    if speed=="":
      speed="N/A"
    else:
      pass                          
    if acc=="":
      acc="N/A"
    else:
      pass
    if brake=="":
      brake="N/A"
    else:
      pass
    if handle=="":
      handle="N/A"
    else:
      pass

    if overall=="":
      overall="N/A"
    else:
      pass

    car_no+=1
    #print(car_no)
    car_dic[car_no]=[car_name,class_name,man_name,vec_fea,price,tspeed,seats,mass,drive,gears,release,speed,acc,brake,handle,overall]
    #print(car_dic)
    car_df=pd.DataFrame.from_dict(car_dic,orient='index',columns=['Car Name','Vehicle Class Name','Manufacturer','Vehicle Features','Price','Top Speed','Number of Seats','Mass','Drive Train','Number of Gears','Release Date','Speed','Acceleration','Brakes','Handling','Overall'])
    print("Data Inserted for ",car_no)
  else:
    car_no+=1    
```
